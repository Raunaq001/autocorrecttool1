```python
import nltk
from autocorrect import Speller
from gramformer import Gramformer
nltk.download('punkt')
spell = Speller()

def autocorrect_and_gramformer_correction(input_text):
    # Spell correction
    corrected_text = " ".join([spell(word) for word in input_text.split()])

    # Grammatical correction using Gramformer
    gramformer = Gramformer()
    gramformer_corrected_text = gramformer.correct(corrected_text)

    return gramformer_corrected_text

# Take user input
user_input = input("Enter a sentence: ")

# Perform autocorrect and grammatical correction using Gramformer
corrected_sentence = autocorrect_and_gramformer_correction(user_input)
print("Corrected Sentence:", corrected_sentence)
```
